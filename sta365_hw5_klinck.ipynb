Provide diagnostic assessments of the performance of your algorithm.*


```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
```

```python
# Part 1
# normal data
n_samples = 200
true_mu = 5
true_sigma = 2
data = np.random.normal(true_mu, true_sigma, size=n_samples)

# Bayesian model using PyMC
with pm.Model() as model:
    # Priors
    theta = pm.Normal("theta", mu=0, sigma=10)  # Mean Prior
    tau = pm.Gamma("tau", alpha=1, beta=1)  # Precision Prior

    # Likelihood
    sigma = pm.Deterministic("sigma", 1 / pm.math.sqrt(tau))  # precision to sd
    likelihood = pm.Normal("obs", mu=theta, sigma=sigma, observed=data)

    # Inference
    trace = pm.sample(2500, return_inferencedata=True, target_accept=0.95)

# Diagnostic Plots
az.plot_trace(trace, compact=False)
plt.show()

# Convergence diagnostics
az.summary(trace)
```

```python
# Part 2

with pm.Model() as model:
    # Priors
    theta = pm.Uniform("theta", lower=-10, upper=10)  # Prior Mean (Non-normal)
    tau = pm.InverseGamma("tau", alpha=2, beta=2)  # Prior precision (Non-gamma)

    # Likelihood
    sigma = pm.Deterministic("sigma", 1 / pm.math.sqrt(tau))  # precision to sd
    likelihood = pm.Normal("obs", mu=theta, sigma=sigma, observed=data)

    # Inference
    trace = pm.sample(2000, return_inferencedata=True, target_accept=0.95)

# Diagnostic Plots
az.plot_trace(trace, compact=False)
plt.show()

# Convergence diagnostics
az.summary(trace)
```

```python
# Part 3

# Bayesian model using PyMC
with pm.Model() as model:
    # Priors
    theta = pm.Laplace("theta", mu=0, b=5)  # Prior Mean (Non-normal)
    tau = pm.Weibull("tau", alpha=2, beta=2)  # Prior Precision (Non-gamma)

    # Likelihood
    sigma = pm.Deterministic("sigma", 1 / pm.math.sqrt(tau))  # precision to sd
    likelihood = pm.Normal("obs", mu=theta, sigma=sigma, observed=data)

    # Inference
    trace = pm.sample(2000, return_inferencedata=True, target_accept=0.95)

# Diagnostic Plots
az.plot_trace(trace, compact=False)
plt.show()

# Convergence diagnostics
az.summary(trace)
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

### How the algo works

The key idea in a slice sampling method is to sample from a region beneath the probability density curve by introducing an auxilary variable.

In this example, the algorithm:

1. **Define the Target Function**: Assume we want to sample from a density function $ f(x) $, which is not necessarily normalized.
   
2. **Introduce an Auxiliary Variable**: Introduce a uniform random variable $ y $:
   \begin{equation}
       y \sim \text{Uniform}(0, f(x))
   \end{equation}
   This defines a horizontal slice at height $ y $, where we must sample a new $ x $.
   
3. **Find the Slice Interval**: Determine the interval $ [a, b] $ where $ f(x) > y $, ensuring:

*   The interval endpoints satisfy $ f(a) < y $ and $ f(b) < y $.
*   The starting point is inside the interval.

4. **Propose a New Sample**: A new $ x' $ is sampled **uniformly** from the interval $ [a, b] $.

5. **Acceptance Step**:

*   If $ f(x') > y $, accept $ x' $ as the new sample.
*   If $ f(x') \leq y $, shrink the interval by adjusting the left or right boundary, then repeat the sampling process.

6. **Repeat Until Acceptance**: Continue this process until a valid sample is found.

### Steps by which slice sampling could be used in place of MH step in a Metropolis within Gibbs algo (where full conditionals are only known up to a normalizing constant)

The Metropolis-within-Gibbs algorithm would be useful when dealing with high-dimensional distributions where each variable's full conditional distribution is known only up to a normalizing constant. In such cases, we replace the standard Metropolis-Hastings (MH) step with a slice sampling step for better efficiency.


```python
import scipy.stats as stats
```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```
